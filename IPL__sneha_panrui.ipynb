```python

# Plotting the distribution of match winners
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/data.csv")
plt.figure(figsize=(10,6))
sns.countplot(y='winner', data=df, order=df['winner'].value_counts().index)
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Number of Wins')
plt.ylabel('Teams')
plt.show()
```

```python

# Plotting the number of matches played in each city
plt.figure(figsize=(10,6))
sns.countplot(y='city', data=df, order=df['city'].value_counts().index)
plt.title('Number of Matches Played in Each City')
plt.xlabel('Number of Matches')
plt.ylabel('City')
plt.show()
```

```python
# Visualizing the result by toss decision
plt.figure(figsize=(6,4))
sns.countplot(x='toss_decision', data=df)
plt.title('Toss Decision Analysis')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

```python


# Correlation heatmap of numerical features
numerical_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(6,6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
